### Introducción
Según la Organización Mundial de la Salud (OMS), el accidente cerebrovascular es la segunda causa principal de muerte a nivel mundial, responsable de aproximadamente el 11% del total de muertes, los que sobreviven corren el riesgo de contar con alguna anomalía como lo es problemas emocionales, problemas del habla, paralitica, perdida de memoria, etc.

Existen varias formas de advertir que se esta sufriendo un accidente cerebrovascular. En la base de datos, cada fila y columna proporciona información relevante sobre el tipo de paciente que cuenta con este problema.

Este conjunto de datos se utiliza para predecir si es probable que un paciente sufra un accidente cerebrovascular en función de los parámetros de entrada como el sexo, edad, enfermedades, tabaquismo, etc. La mayoría de los accidentes cerebrovasculares se producen en personas de mayor edad, y la mayoría ocurre sin advertencia previa.

### Descripción de datos
* **Id:** Identificación única.
* **Género:** El tipo de dato tipo texto. Masculino, Femenino u otro.
* **Edad:** Dato tipo entero. Edad del paciente.
* **Característica binaria de hipertensión:** Tipo entero. Si el paciente cuenta con hipertensión = 1, si no cuenta con hipertensión = 0.
* **Característica binaria de enfermedad cardíaca:** Tipo entero. Si el paciente cuenta con una enfermedad cardíaca = 1, si el paciente no cuenta con una enfermedad cardíaca = 0.
* **¿El paciente ha estado casado alguna vez?:** Dato tipo texto. Estatus marital (cierto u falso).
* **Tipo de trabajo del paciente:** Dato tipo texto. Tipo de empleo (Privado, Cuenta propia, etc.)
* **Tipo de residencia del paciente:** Dato tipo texto. Rural o Urbano.
* **Nivel medio de glucosa en sangre:** Dato tipo flotante. Nivel medio de glucosa en sangre de una persona.
* **Índice de masa corporal:** Dato tipo flotante. Índice de masa corporal de una persona.
* **Estatud de fumadores:** Dato tipo texto. "anteriormente fumado", "nunca fumado", "fuma"o "Desconocido" *
* **Accidente cerebrovascular:** Dato tipo entero. Se considera 1 si el paciente tuvo un accidente cerebrovascular o 0 si no

### <font color='#6e7681'>Técnicas para predecir futuros casos de ACV:</font>
Primero, empleando la función **LabelEncoder**, preparamos los datos, esta función transforma datos categóricos en númericos, esto lo hacemos desde un comienzo, ya que nos servirá en cualquier técnica:

```python
from sklearn import preprocessing
lbl=preprocessing.LabelEncoder()
cf['Sexo']=lbl.fit_transform(cf['Sexo'])
cf['Estado respecto a si fuma']=lbl.fit_transform(cf['Estado respecto a si fuma'])
```

### Máquinas de soporte vectorial
Una vez realizado lo anterior, comenzamos con la primera técnica a emplear, que son las métricas de evaluación, comenzaremos empleando una función de la librería sklearn, la cual es:
* **test_train_split:** Esta función nos permite dividir un dataset en dos bloques, típicamente bloques destinados al entrenamiento y validación del modelo.

```python
from sklearn.model_selection import train_test_split
```

Para comenzar, separamos las variables predictoras de la variable de respuesta:

```python
aux1=list(cf.columns[2:7])
predictores=cf[aux1]
respuesta=cf['stroke']
```

Se crean las variables de entrenamiento y las variables de prueba:

```python
X_entrenamiento, X_test, y_entrenamiento, y_test = train_test_split(predictores, respuesta,test_size=0.25,random_state=7)
```

Después de esto, procederemos a llamar a la función **StandardScaler** de la librería sklearn, la cual nos ayuda en la clasificación de vectores de soporte:

```python
from sklearn.svm import SVC
```

Después de toda la preparación de variables, ahora se puede emplear la técnica de máquinas de soporte de vectorial, empleando en el kernel del SCV *sigmoid*, que se refiere a una función sigmoide, la cual comúnmente se refiere al caso partícular de la función logística:

```python
clasificador1=SVC(kernel="sigmoid")
clasificador1.fit(X_entrenamiento, y_entrenamiento)
yp1=clasificador1.predict(X_test)
#Ayuda de: https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html
```

Ahora, empleando la función **confusion_matrix**, crearemos la matriz de confusión de la técnica:

```python
from sklearn.metrics import confusion_matrix
cm1=confusion_matrix(y_test,yp1)
print("Matriz de confusión:")
print(cm1)
print("\nHabrá",yp1.sum(),"casos nuevos de ACV.")
#Ayuda de: https://scikit-learn.org/stable/modules/generated/sklearn.metrics.confusion_matrix.html
```

La presición de la técnica en base a las variables de entrenamiento es:

```python
e=clasificador1.score(X_entrenamiento, y_entrenamiento)
print("Precisión: ",round(e*100,2),"%")
```

La presición de la técnica en base a las variables de prueba es:

```python
pb=clasificador1.score(X_test, y_test)
print("Precisión: ",round(pb*100,2),"%")
```

### Árboles de decisión
La segunda técnica a emplear son los árboles de decisión, ya que este nos muestra de una manera visual, como se presenta un ACV según los padecimientos o comorbilidades que se tienen, emplearemos las siguientes funciones:
* **DecisionTreeClassifier:** En esta función clasifica los árboles de decisión.

```python
from sklearn.tree import DecisionTreeClassifier
```

Reutilizando algunas variables de la técnica anterior encontramos lo siguiente:

```python
clasificador2=DecisionTreeClassifier(criterion="entropy",random_state=0)
clasificador2.fit(X_entrenamiento,y_entrenamiento)
yp2=clasificador2.predict(X_test)
cm2=confusion_matrix(y_test,yp2)
print("Matriz de confusión:")
print(cm2)
print("\nHabrá",yp2.sum(),"casos nuevos de ACV.")
#Ayuda de: https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html
```

La precisión del árbol de decisión con las variables de entrenamiento es:

```python
et=clasificador2.score(X_entrenamiento, y_entrenamiento)
print("Precisión: ",round(et*100,2),"%")
```

La precisión del árbol de decisión con las variables de prueba es:

```python
ts=clasificador2.score(X_test, y_test)
print("Precisión: ",round(ts*100,2),"%")
```

Procederemos a buscar cuanta profundidad tiene nuestro árbol de decisión antes de visualizarlo:

```python
clasificador2.tree_.max_depth
```

Al ser de 22 líneas es muy enorme, por lo que lo reduciremos a 4 y volveremos a crearlo:

```python
clasificador2=DecisionTreeClassifier(max_depth=4)
clasificador2.fit(X_entrenamiento,y_entrenamiento)
print("Precisión en variables entrenamiento: ",round(clasificador2.score(X_entrenamiento, y_entrenamiento)*100,2),"%")
print("Precisión en variables prueba: ",round(clasificador2.score(X_test, y_test)*100,2),"%")
```
