```python
import pandas as pd
import numpy as np
import json
df = pd.read_csv("C:/Users/Daniela/Desktop/Daniela/Facultad/7mo sem/Mineria/AVANCE PIA/BASE 1/data_mmda_traffic_spatial.csv")
```

Ahora, como ya mencionamos el enfoque que le queremos dar a esta base de datos, tuvimos que retirar ciertas columnas que creemos inecesarias para nuestro proyecto. Dejamos la fecha para asi despues generar un resgistro de los accidentes que hubo entre los intervalos de tiempo, la ciudad y locación para después generar el estudio de la frecuencia de incidentes en Manila, la latitud y longitud para ser mas exactos donde se genero el accidente, y el tipo de automovil y vehiculos involucrados, para saber que tipo de población es la que mas frecuenta a los accidentes.

```python
df.rename(columns={'Date':'Fecha','City':'Ciudad','Location':'Locación','Latitude':'Latitud','Longitude':'Longitud','Type':'Tipo de automovil','Involved':'Vehiculos Involucrados'},inplace=True)
#Modifica los nombres de las columnas, con ayuda de: https://www.analyticslane.com/2019/05/06/como-cambiar-el-nombre-de-las-columnas-en-pandas/
cf=df.drop(['Time','Direction','High_Accuracy','Lanes_Blocked','Tweet','Source'],axis=1) #Guarda la base que necesitamos en otra variable
#Con ayuda de: https://www.analyticslane.com/2019/03/25/como-eliminar-columnas-y-filas-en-un-dataframe-pandas/
cf
```

```python
df.isnull().any()
```

```python
df.dtypes #Defininos el tipo de datos que manejamos en nuestr base
```

```python
cf.isnull().sum() #Comprobando datos faltantes
```

```python
cf['Ciudad'].fillna(0,inplace=True)
cf['Locacion'].fillna(0,inplace=True)
cf['Tipo de automovil'].fillna(0,inplace=True)
cf['Vehiculos Involucrados'].fillna(0,inplace=True)
```

```python
cf.to_csv("NuevaBaseAccidentes.csv",index=False)
#Lugar donde obtuvimos la ayuda: https://aprendeia.com/limpieza-y-procesamiento-de-datos-con-codigo-en-python/
```

Existen varias formas de advertir que se esta sufriendo un accidente cerebrovascular. En la base de datos, cada fila y columna proporciona información relevante sobre el tipo de paciente que cuenta con este problema.

Este conjunto de datos se utiliza para predecir si es probable que un paciente sufra un accidente cerebrovascular en función de los parámetros de entrada como el sexo, edad, enfermedades, tabaquismo, etc. La mayoría de los accidentes cerebrovasculares se producen en personas de mayor edad, y la mayoría ocurre sin advertencia previa.
